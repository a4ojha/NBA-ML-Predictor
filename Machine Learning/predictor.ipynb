```python
from datetime import datetime

def convert_time(time_str):
    # Check if time is PM and convert to 24-hour format
    if time_str.endswith('p'):
        # Remove the 'p' suffix and parse the time
        time_obj = datetime.strptime(time_str[:-1], '%I:%M')
        # Add 12 hours if the time is not 12 PM to convert to 24-hour format
        time_obj = time_obj if time_obj.hour == 12 else datetime(time_obj.year, time_obj.month, time_obj.day, time_obj.hour + 12, time_obj.minute)
    elif time_str.endswith('a'):
        # Remove the 'a' suffix and parse the time, no need to add hours as 'a' indicates AM
        time_obj = datetime.strptime(time_str[:-1], '%I:%M')
        # Convert 12 AM to 00 hours
        time_obj = time_obj if time_obj.hour != 12 else datetime(time_obj.year, time_obj.month, time_obj.day, 0, time_obj.minute)
    else:
        # If no 'a' or 'p' suffix, assume 24-hour format already
        time_obj = datetime.strptime(time_str, '%H:%M')
    
    # Return the time in 24-hour format
    return time_obj.strftime('%H:%M')
```

```python
games["Start (ET)"] = games["Start (ET)"].apply(convert_time)
games.rename(columns={'Start (ET)': 'Start'}, inplace=True)
```

```python
# Turns 19:00 (7:00pm) to 19
games["hour"] = games["Start"].str.replace(":.+", "", regex=True).astype("int")
```

```python
games["day_code"] = games["Date"].dt.dayofweek
```

```python
games
```

```python
# Set up a target to predict
# Target 1: Whether the team won or not
games["target"] = (games["W/L"] == "W").astype("int")
```

```python
games
```

```python
# This library can pick up non-linearities in the data
from sklearn.ensemble import RandomForestClassifier
```

```python
# A RandomForest is a series of decision trees, but each decision tree has slightly different parameters

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)

# n_estimators = number of individual decision trees we want to train. Higher the nunmber, longer the algo takes to run, but more accurate
# min_samples_split = number of samples we want in the leaf of a decision tree. Higher the number, less likely we are to overfit, but lowered accuracy on training data
# random_state = If random forest is ran multiple times, we'll get the same results (as long as the data is the same)
```

```python
# Train based on all matches from before mid 2023
train = games[games["Date"] < '2023-06-01']
```

```python
# Test on games AFTER mid 2023
test = games[games["Date"] > "2023-06-01"]
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
# .fit() will train a Random Forect model with 'predictors', to guess 'target' (which is 0 for a loss, 1 for a win)
rf.fit(train[predictors], train["target"])
```

```python
# Generate predictions using .predict() by passing in test data
preds = rf.predict(test[predictors])
```

```python
# Now, determine accuracy of model.

# Metric 1: If win is predicted, what percentage of the time did the team ACTUALLY win. and vice versa for losses
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
# The output of 0.57 means that our predictions were 57% accurate... not great
acc
```

### Improve precision with rolling averages

```python
# Create one dataframe for every team in the NBA
grouped_games = games.groupby("Team")
```

```python
# Data of all Raptors games from 2023 and 2024 season (up to Jan. 28, 2024)
group = grouped_games.get_group("Boston Celtics")
```

```python
group
```

```python
# Parameters: Team, Set of columns to compute avgs for, set of new columns to assign rolling avgs to
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
# Want to create rolling averages for the following in-game metrics
cols = ["FG", "FGA", "3P", "3PA", "3P%", "FT", "FTA", "FT%", "ORB", "TRB", "AST", "STL", "BLK", "TOV", "PF"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
games_rolling = games.groupby("Team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
games_rolling
```

```python
games_rolling = games_rolling.droplevel("Team")
```

```python
games_rolling
```

```python
games_rolling.index = range(games_rolling.shape[0])
```

```python
games_rolling.shape
```

```python
# Makes it easy for us to continue to iterate on our algorithm (removes the need to re-train)
def make_predictions(data, predictors):
    train = data[data["Date"] < "2023-06-01"]       # Train data pre-2023 Jan
    test = data[data["Date"] > "2023-06-01"]        # Test on data post-2023 Jan
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], prediction=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(games_rolling, predictors + new_cols)
```

```python
# Accuracy: went from 57% to 74%.. pretty good.
precision
```
